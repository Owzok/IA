## Linear Regression

```python
from sklearn import linear_model
import pandas as pd
import numpy as np
                    # Load data
data = pd.read_csv("../../Data/wine_quality.csv", sep = ";")
data
```

#### | Values in the train and test datasets must be numeric and in numpy arrays |

```python
# Train dataset
x_train = data.iloc[:-10:,0:-1]
x_train
```

```python
y_train = data.iloc[:-10:,-1]
y_train
```

```python
x_test = data.iloc[-10::,0:-1]
x_test
```

```python
y_test = data.iloc[-10::,-1]
y_test = np.array(y_test)
y_test
```

```python
linear = linear_model.LinearRegression()
linear.fit(x_train, y_train)
linear.score(x_train, y_train)

print("Coefficient:", linear.coef_)
print("Intercept:", linear.intercept_)

predicted = linear.predict(x_test)
predicted
```

```python
def check_precision(test, predicted):
    correct = 0
    for i in range(len(test)):
        if round(test[i]) == round(predicted[i]):
            correct += 1
    return (correct/len(test))*100

check_precision(y_test, predicted)
```
